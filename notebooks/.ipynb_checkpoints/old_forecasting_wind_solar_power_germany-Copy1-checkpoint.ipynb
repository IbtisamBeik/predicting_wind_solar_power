```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Read power generation data
production = pd.read_csv("./data/time_series_60min_singleindex_filtered.csv",
                        parse_dates=[0], index_col=0)
```

```python
production
```

```python
production.info()
```

```python
# just keep the columns relevant for the project, i.e. the total power generation and capacity
columns = ["DE_solar_capacity", 
           "DE_solar_generation_actual", 
           "DE_wind_capacity",
           "DE_wind_generation_actual"]

production = production[[col for col in columns]]
production = production.rename(columns={"DE_solar_capacity": "solar_capacity",
                                        "DE_solar_generation_actual": "solar_generation_actual",
                                        "DE_wind_capacity": "wind_capacity",
                                        "DE_wind_generation_actual": "wind_generation_actual"})
production
```

```python
# When plotting solar and wind power generation over time (i.e. index), the following error kept on popping up:
# plot float() argument must be a string or a number, not 'Timestamp'
# It appears that the matplotlib converters were deregistered.
# The following code circumvents this issue:
pd.plotting.register_matplotlib_converters()
```

```python
# Exploratory Data Analysis: create plot for wind power generation
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(production.index,
        production['wind_generation_actual'],
        color="b");
ax.set_title("Wind generation in Germany in MW");
ax.set_xlim((pd.to_datetime("2015-01-01"), pd.to_datetime("2017-01-01")));
ax.set_ylim((0,35000));
ax.set_frame_on(False);

# save plot
# plt.savefig("figs/wind.png", dpi=200)
```

```python
# Exploratory Data Analysis: create plot for solar power generation
fig, ax = plt.subplots(figsize=(8,5))

ax.plot(production.index,
        production['solar_generation_actual'],
        color="OrangeRed");
ax.set_title("Solar generation in Germany in MW");
ax.set_xlim((pd.to_datetime("2015-01-01"), pd.to_datetime("2017-01-01")));
ax.set_ylim((0,27000));
ax.set_frame_on(False);

# save plot
# plt.savefig("figs/solar.png", dpi=200)
```

```python
# Read weather data
weather = pd.read_csv("./data/weather_data_filtered.csv",
                     parse_dates=[0], index_col=0)
```

```python
weather
```

```python
weather.info()
```

```python
weather.isna().sum()
```

```python
# Since you consider wind and solar generation at the national level, we drop the columns for regional data (NUTS zones: https://ec.europa.eu/eurostat/documents/345175/7451602/2016-NUTS-2-map-DE.pdf)
weather = weather[["DE_windspeed_10m","DE_temperature","DE_radiation_direct_horizontal", "DE_radiation_diffuse_horizontal"]]
weather
```

```python
weather.isna().sum()
```

```python
# The number and nature of features for weather data is believed not to be comprehensible enough.
# Another dataset found on https://www.renewables.ninja/ could complete the original weather dataset since it contained information on:
# precipitation in mm/hour (PRECTOTLAND in MERRA-2) 
# temperature at 2 metres above ground in degrees C (T2M in MERRA-2)
# irradiance at ground level (_surface) in W/m^2 (SWGDN in MERRA-2)
# irradiance at top of atmosphere (_toa) in W/m^2 (SWTDN in MERRA-2)
# snowfall in mm/hour (PRECSNOLAND in MERRA-2)
# snow_mass in kg/m^2 (SNOMAS in MERRA-2)
# cloud_cover as a fraction [0,1] (CLDTOT in MERRA-2)
# air_density at ground level in kg/m^3 (RHOA in MERRA-2)
weather_2 = pd.read_csv("./data/ninja_weather_country_DE_merra-2_land_area_weighted.csv",
                        parse_dates=[0], index_col=0)
weather_2
```

```python
weather_2 = (weather_2
             .loc[(weather_2.index.year == 2015) | (weather_2.index.year == 2016)]
             .tz_localize('utc'))
weather_2
```

```python
weather = pd.merge(weather, weather_2, how='inner', left_index=True, right_index=True)
```

```python
# plot and compare the two temperature features to see why they seem to differ
# plt.scatter(weather["DE_temperature"], weather["temperature"])
sns.regplot(data=weather,
            x="DE_temperature",
            y="temperature");
```

```python
weather[["DE_temperature", "temperature"]].corr()
```

```python
# Since there's no information on how the mean temperature for Germany was calculated for both temperature features
# And since their values are not significantly different
# We can drop any of the two features for further analysis
weather = (weather
           .drop(columns= "DE_temperature")
           .rename(columns={"DE_windspeed_10m": "windspeed_10m", 
                            "DE_radiation_direct_horizontal": "radiation_direct_horizontal",
                            "DE_radiation_diffuse_horizontal": "radiation_diffuse_horizontal"}))
weather
```

```python
# import necessary modules
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# instantiate LinearRegression
lr = LinearRegression()
```

```python
wind = combined[['wind_generation_actual', 'wind_capacity', 'windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']]
# X_wind = wind[['windspeed_10m', 'temperature', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal']]
X_wind = wind[['wind_capacity', 'windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']]
y_wind = wind['wind_generation_actual']
wind
```

```python
# we implement a 5-fold CV procedure
scores_wind = cross_val_score(lr, X_wind, y_wind, cv=5)
print(scores_wind, "\naverage =", np.mean(scores_wind))
```

```python
solar = combined[['solar_generation_actual', 'solar_capacity', 'windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']]
X_solar = solar[['solar_capacity','windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']]
# X_solar = solar[['windspeed_10m', 'temperature', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal']]
y_solar = solar['solar_generation_actual']
solar
```

```python
# we implement a 5-fold CV procedure
scores_solar = cross_val_score(lr, X_solar, y_solar, cv=5)
print(scores_solar, "\naverage =", np.mean(scores_solar))
```

```python
lr = LinearRegression()
lr.fit(X_solar, y_solar)
lr.intercept_, lr.coef_
```

```python
# Let's test the Linear Regression to forecast solar power generation for 2014
weather_2014_2 = pd.read_csv("./data/ninja_weather_country_DE_merra-2_land_area_weighted.csv",
                           parse_dates=[0], index_col=0)
weather_2014_2 = (weather_2014_2
                  .loc[(weather_2014_2.index.year == 2014)]
                  .tz_localize('utc'))

weather_2014 = pd.merge(weather_2014, weather_2014_2, how='inner', left_index=True, right_index=True)

weather_2014 = (weather_2014
                .drop(columns= "DE_temperature")
                .rename(columns={"DE_windspeed_10m": "windspeed_10m", 
                                 "DE_radiation_direct_horizontal": "radiation_direct_horizontal",
                                 "DE_radiation_diffuse_horizontal": "radiation_diffuse_horizontal"}))
weather_2014
```

```python
# Read power generation data for 2014
production_2014 = pd.read_csv("./data/time_series_60min_singleindex_filtered_2014.csv",
                              parse_dates=[0], index_col=0)
production_2014 = production_2014[[col for col in columns]]
production_2014 = production_2014.rename(columns={"DE_solar_capacity": "solar_capacity",
                                        "DE_solar_generation_actual": "solar_generation_actual",
                                        "DE_wind_capacity": "wind_capacity",
                                        "DE_wind_generation_actual": "wind_generation_actual"})
production_2014
```

```python
# merge production and weather DataFrames
combined_2014 = pd.merge(production_2014, weather_2014, how='inner', left_index=True, right_index=True)
combined_2014
```

```python
combined_2014.isna().sum()
```

```python
solar_2014 = combined_2014.dropna()
# combined_2014.loc[combined_2014["solar_generation_actual"] is nan]
#combined_2014 = combined_2014.loc[combined_2014['solar_generation_actual']isna]
```

```python
train_X = X_solar 
train_y = y_solar
test_X = solar_2014[['wind_capacity', 'windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']]
test_y = solar_2014['solar_generation_actual']
# r2 score, mse (mean squared error), rmse (root mean squared error).
from sklearn.metrics import r2_score

lr.fit(train_X, train_y)
predictions = lr.predict(test_X)
r2 = r2_score(test_y, predictions)
print(f"The R2 score of the model is {r2}")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# to work with mse & rmse, we need to standardize the datasets
std = StandardScaler()
solar_cols = ['solar_generation_actual', 'solar_capacity', 'windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']
scaled_fit_solar = std.fit(solar[solar_cols])
scaled_solar_train = pd.DataFrame(scaled_fit_solar.transform(solar[solar_cols]), columns=solar_cols)
scaled_solar_train
```

```python
solar_cols = ['solar_generation_actual', 'solar_capacity', 'windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']
scaled_fit_solar_2014 = std.fit(combined_2014[solar_cols])
scaled_solar_test = pd.DataFrame(scaled_fit_solar.transform(combined_2014[solar_cols]), columns=solar_cols)
scaled_solar_test = scaled_solar_test.dropna()
scaled_solar_test
```

```python
from sklearn.metrics import mean_squared_error

train_X_scaled = scaled_solar_train[['windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']]
train_y_scaled = scaled_solar_train['solar_generation_actual']

test_X_scaled = scaled_solar_test[['windspeed_10m', 'radiation_direct_horizontal', 'radiation_diffuse_horizontal', 'irradiance_surface', 'irradiance_toa', 'precipitation', 'snowfall', 'cloud_cover', 'air_density']]
test_y_scaled = scaled_solar_test['solar_generation_actual']

lr.fit(train_X_scaled, train_y_scaled)
predictions = lr.predict(test_X_scaled)
mse = mean_squared_error(test_y_scaled, predictions)
print(f"The mean square error of the model is {mse}")
```
